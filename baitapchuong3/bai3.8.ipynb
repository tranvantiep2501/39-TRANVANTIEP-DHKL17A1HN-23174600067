```python
import numpy as np 
```

```python
height = []
position = []

file_path_h = 'heights.txt'
file_path_p = 'positions.txt'

with open(file_path_h, 'r', encoding='utf-8-sig') as file : 
    for line in file : 
        height.append(line.strip())

with open(file_path_p, 'r', encoding='utf-8-sig') as file : 
    for line in file : 
        position.append(line.strip())

height_str = ''.join(height) 
position_str = ''.join(position)

height = [float(x) for x in height_str.split(',')]
position = [x for x in position_str.split(',')]

print(height)
print(position)
```

```python
#  1 : to array v xem kiu d liu ca array  : 
np_positions = np.array(position)
print("Mng : ", np_positions)
print("kiu d liu l : ", np_positions.dtype)

np_heights = np.array(height)
print("Mng : ", np_heights)
print("kiu d liu l : ", np_heights.dtype)
```

```python
#  2 : Chiu cao trung bnh ca cc GK : 
heights_gk = np_heights[np_positions == 'GK']
avg_height_gk = np.mean(heights_gk)
print("\nChiu cao trung bnh ca cc GK:", avg_height_gk)
```

```python
#  3 : Chiu cao trung bnh ca cc v tr khc : 
heights_other = np_heights[np_positions != 'GK']
avg_height_other = np.mean(heights_other)
print("Chiu cao trung bnh ca cc v tr khc:", avg_height_other)
```

```python
#  4 : To mng d liu : 
players = np.zeros(len(position), dtype=[('position', 'U5'), ('height', 'f4')])
players['position'] = np_positions
players['height'] = np_heights
```

```python
#  5 : v tr c chiu cao cao nht v chiu cao thp nht : 
players_sorted = np.sort(players, order='height')
highest = players_sorted[-1]
lowest = players_sorted[0]
print("\nV tr c chiu cao cao nht:", highest['position'], "vi chiu cao:", highest['height'])
print("V tr c chiu cao thp nht:", lowest['position'], "vi chiu cao:", lowest['height'])
```
