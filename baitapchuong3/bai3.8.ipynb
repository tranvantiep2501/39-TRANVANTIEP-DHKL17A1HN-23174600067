```python
import numpy as np 
```

```python
height = []
position = []

file_path_h = 'heights.txt'
file_path_p = 'positions.txt'

with open(file_path_h, 'r', encoding='utf-8-sig') as file : 
    for line in file : 
        height.append(line.strip())

with open(file_path_p, 'r', encoding='utf-8-sig') as file : 
    for line in file : 
        position.append(line.strip())

height_str = ''.join(height) 
position_str = ''.join(position)

height = [float(x) for x in height_str.split(',')]
position = [x for x in position_str.split(',')]

print(height)
print(position)
```

```python
# Ý 1 : tạo array và xem kiểu dữ liệu của array đó : 
np_positions = np.array(position)
print("Mảng : ", np_positions)
print("kiểu dữ liệu là : ", np_positions.dtype)

np_heights = np.array(height)
print("Mảng : ", np_heights)
print("kiểu dữ liệu là : ", np_heights.dtype)
```

```python
# ý 2 : Chiều cao trung bình của các GK : 
heights_gk = np_heights[np_positions == 'GK']
avg_height_gk = np.mean(heights_gk)
print("\nChiều cao trung bình của các GK:", avg_height_gk)
```

```python
# ý 3 : Chiều cao trung bình của các vị trí khác : 
heights_other = np_heights[np_positions != 'GK']
avg_height_other = np.mean(heights_other)
print("Chiều cao trung bình của các vị trí khác:", avg_height_other)
```

```python
# ý 4 : Tạo mảng dữ liệu : 
players = np.zeros(len(position), dtype=[('position', 'U5'), ('height', 'f4')])
players['position'] = np_positions
players['height'] = np_heights
```

```python
# ý 5 : vị trí có chiều cao cao nhất và chiều cao thấp nhất : 
players_sorted = np.sort(players, order='height')
highest = players_sorted[-1]
lowest = players_sorted[0]
print("\nVị trí có chiều cao cao nhất:", highest['position'], "với chiều cao:", highest['height'])
print("Vị trí có chiều cao thấp nhất:", lowest['position'], "với chiều cao:", lowest['height'])
```
