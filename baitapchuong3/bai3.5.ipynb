```python
height = []
weight = []

file_path_h = 'heights_1.txt'
file_path_w = 'weights_1.txt'

with open(file_path_h, 'r', encoding='utf-8-sig') as file : 
    for line in file : 
        height.append(line.strip())

with open(file_path_w, 'r', encoding='utf-8-sig') as file : 
    for line in file : 
        weight.append(line.strip())

height_str = ''.join(height)
weight_str = ''.join(weight)

height = [float(x) for x in height_str.split(',')]
weight = [float(x) for x in weight_str.split(',')]

print(height)
print(weight)
```

```python
# Ã 1: Táº¡o máº£ng array cho height : 
arr_height = np.array(height)
arr_height
```

```python
# Ã 2: Taá» máº£ng array cho weight : 
arr_weight = np.array(weight)
arr_weight
```

```python
# Ã 3: inch sang m lÃ  0.0254 : 
arr_height_m = arr_height * 0.0254
arr_height_m
```

```python
# Ã 4: pound sang kg lÃ  0.453592 : 
arr_weight_kg = arr_weight * 0.453592
arr_weight_kg
```

```python
# Ã 5: TÃ­nh BMI : 
arr_bmi = arr_weight_kg / (arr_height_m * arr_height_m)
arr_bmi
```

```python
# Ã½ 6: GiÃ¡ trá» á» vá» trÃ­ index = 50 : 
arr_weight_kg[50]
```

```python
# Ã 7: Pháº§n tá»­ cÃ³ vá» trÃ­ index tá»« 100 Äáº¿n 110 : 
arr_height_m_100 = arr_height_m[100:111]
arr_height_m_100
```

```python
# Ã 8: Cáº§u thá»§ bÃ³ng chÃ y cÃ³ bmi < 21 : 
arr_bmi_if = np.where(arr_bmi < 21)[0] # hÃ m np.where() tÃ¬m vá» trÃ­ index dá»±a trÃªn Äiá»u kiá»n cÃ³ trong ngoáº·c 
arr_bmi_21 = arr_bmi[arr_bmi_if]
arr_bmi_21
```

```python
# Ã½ 9: Chiá»u cao trung bÃ¬nh vÃ  cÃ¢n náº·ng trung bÃ¬nh cá»§a cÃ¡c cáº§u thá»§ : (DÃ¹ng hÃ m np.mean()Äá» kiá»m tra sá» trung bÃ¬nh cÃ³ trong array)
arr_height_medium = np.mean(arr_height_m)
arr_weight_medium = np.mean(arr_weight_kg)
print(arr_height_medium, arr_weight_medium)
```

```python
# Ã½ 10 : Chiá»u cao vÃ  cÃ¢n náº·ng lá»n nháº¥t cá»§a cÃ¡c cáº§u thá»§ : 
arr_height_max = np.max(arr_height_m)
arr_weight_max = np.max(arr_weight_kg)
print(arr_height_max, arr_weight_max)
```

```python
# Ã½ 11 : CHiá»u cao vÃ  cÃ¢n náº·ng nhá» nháº¥t cá»§a cÃ¡c cáº§u thá»§ : 
arr_height_min = np.min(arr_height_m)
arr_weight_min = np.min(arr_weight_kg)
print(arr_height_min, arr_weight_min)
```
